## Data Transform from wide data to a tidy format that we are presenting:

So what exactly does the current problem have? well there is a few that should be addressed:

- Monthly plyaer counts is basically spread across multiple columns
- the time sereis analysis is difficult for one to understand 
- There is an issue with creating the data visualization and would really be difficult overtime. 

How will my work benefit if we tranfrom it to a 'tidy' format:
- With each variable that is presented they will have their own column and basically makes it look clean overall
- with each observation we will that into a row
- and each of the unti that we are working with will have a table 

```python
# over here weill be loading the steam dataset from the csv file that is provided
df = pd.read_csv("steam_data.csv")

# It's going to be displaying the first few rows that is presented
print("Loaded dataset from CSV:")
df.head()
```

```python
# convert the data into a tidy and when we use melt method basically with what is being occured is that its basically going to go to a long format and its uses for data manipulation and having a better idea with the analysis that we are going to work with. 

df_tidy = pd.melt(
    df,
    id_vars=["game_id", "game_name", "developer", "release_year", "genre", "early_access", "average_review_score", "price_usd"],
    value_vars=["jan_2023_players", "feb_2023_players", "mar_2023_players", "apr_2023_players", "may_2023_players"],
    var_name="month",
    value_name="player_count"
)

# basically just aclean up the column and that is by remocing the players with using the method of str.replace and makes it more clean 
df_tidy["month"] = df_tidy["month"].str.replace("_players", "").str.replace("_", " ").str.title()

print("\nTransformed 'tidy' format data:")
df_tidy.head(10)
```

## The data analytical approach we are taking 

So once I have managed to transform the data that is being provided I will basically look for the key aspects that we are encountering with the engagement

### 1. Genre performance

First thing that I will be reviewing is basically the game generes and how they would typically perform when it comes down to counts and understanding the trends and by doing this it will help us a lot and make us see what genre tend to maintain and consistent player which cold be useful for publishers 

```python
# what is going to be occured here is that we are going to calculate the avg of the plyaer count which comes with the genre and the month that is being presented 

genre_avg = df_tidy.groupby(["genre", "month"])["player_count"].mean().reset_index()
print("\nAverage player counts by genre and month:")
genre_avg
```

## 2. Understanding the player growth analysis 

Now let's try to calcualte the growth and the play counts. By getting to understand the calcualtion and the change that is being occured with the players from the month January to the month of May 2023, What is basically going on is basically there is a popularity and understand with what is delcing in the market and basically by doing this. I twould give us a much better insight overall for these types of games. 

```python
# what we are goingto do here basically try to understand and calculating the plyaer change that is presented 

game_trends = df_tidy.pivot_table(index="game_name", columns="month", values="player_count").reset_index()
# overe here will focus on the column and basically confirm the names
print("Available columns:", game_trends.columns.tolist())

# we will be calcualting the growth of the metrics 
game_trends["player_growth"] = game_trends["May 2023"] - game_trends["Jan 2023"]
game_trends["growth_percent"] = ((game_trends["May 2023"] - game_trends["Jan 2023"]) / game_trends["Jan 2023"]) * 100
game_trends = game_trends.sort_values("growth_percent", ascending=False)

print("\nGame growth from January to May 2023:")
game_trends[["game_name", "Jan 2023", "May 2023", "player_growth", "growth_percent"]]
```

## 3. The attribute that is provided and the correlations

What I will be doing nis now that I will basically analyze and try to understand the relationship that is being occured between the game attricutes and that comes with the release year, the scores, and by doing this it's going to reveal crucical factors that shoudl be reviewed and understanding the longeivity of the game. 

```python
# what are we gonna do is basically we are going to be merging the data of the game attributes 
growth_analysis = pd.merge(
    game_trends[["game_name", "growth_percent"]],
    df[["game_name", "release_year", "average_review_score", "price_usd"]].drop_duplicates(),
    on="game_name"
)

print("\nGrowth analysis with game attributes:")
growth_analysis

# Now we will se a correlation that is presented with the growth and other factors taht come with 
print("\nCorrelation between growth percentage and game attributes:")
correlations = growth_analysis[["growth_percent", "release_year", "average_review_score", "price_usd"]].corr()
correlations["growth_percent"]
```

## Visualizations

### Monthly player count

One of the thing that I really love of visualizing my data with matplotlib or in general is that it tells us a story and it's easy for the viewer to consume which is what I like to include in my work, now I will be in putting the code and make my graph come to life! So this overall is basically what is gonna do is help us with telling a story with the data that we are working and makes it easier for us to understand with the trend that is going. 

```python
plt.figure(figsize=(12, 6))
for game_id, game_data in df_tidy.groupby("game_id"):
    game_name = game_data["game_name"].iloc[0]
    plt.plot(game_data["month"], game_data["player_count"], marker='o', label=game_name)

plt.title("Monthly Player Counts by Game (Jan-May 2023)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Player Count", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

## The player growth comparision that is being made

This bar that I am presenting will help see with which game is actually gowing and which game is declining and a game could be declining because of hte possibility and basically the ability to be able to compare the performance with diff titles 

```python
plt.figure(figsize=(10, 6))
plt.bar(game_trends["game_name"], game_trends["growth_percent"], color='skyblue')
plt.title('Player Growth Percentage (Jan to May 2023)', fontsize=14)
plt.xlabel('Game', fontsize=12)
plt.ylabel('Growth (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## The player count that is being presented by the genre

Now for this visualization what is being occured is how the the genres is getting to be performed in terms of hwow the engagement is be observed 

```python
plt.figure(figsize=(12, 6))
for genre, genre_data in df_tidy.groupby("genre"):
    genre_avg = genre_data.groupby("month")["player_count"].mean()
    plt.plot(genre_avg.index, genre_avg.values, marker='o', label=genre)

plt.title("Average Player Counts by Genre (Jan-May 2023)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Player Count", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
