## Now it's time to transform the data from a wide data to a tidy format

The the dataset that I am currently working with clearly has sme issues that would really make certain analysis challenging:

Those could be

- We see that the the metrics of the performance is being spread across multiple columns that is presented in our work
- Getting to analzyze the realtionships between teh metrics would be very complex and mroe steps would be required
- When it comes down to visualizing the data with the different groups that is presented it becomes more difficult

The steps of how to transform to a tidy format and what can it benefit or bring to my analysis that I will be putting out:

- Tidying the data will help by having each of the variable having it's own column that we can work with
- we will have each of the observation that will be forming a row
- the grouping and having the aggregation will be a lmore mroe straightforward 

This transformantion will of course give us a better chance to analyze the data and show us on the ability to:

- Understand the metrics  of the perfromance from each different fighter group
- It will also help us with the realtionship and getting to understand the attributues and with the performance metrics
- With being more informational and provide visuals that can help us with the information that is being provided.

```python
#  Like we have done in our last two data sets let's actually focus and categoritze on our metrics and what we are measuring 

fighter_attributes = ['name', 'age', 'height', 'weight', 'reach', 'stance', 'weight_class']
performance_metrics = ['wins', 'losses', 'SLpM', 'sig_str_acc', 'SApM', 'str_def', 
                       'td_avg', 'td_acc', 'td_def', 'sub_avg']

# now we whe will be doing is basically creating a data set the same way we have been doing using the melt method. 

df_tidy = pd.melt(
    df,
    id_vars=fighter_attributes,
    value_vars=performance_metrics,
    var_name='metric',
    value_name='value'
)

# we are now going to be displaying the first few rows of the data set "tidy" that is being presented 

print("\nTransformed 'tidy' format data:")
df_tidy.head(10)
```

This tranformation will change the data strucute in a massive way:

1. **Going from a wide set to a long tidy set** This is going to help us because why have 10 separate metric across the board, in this scneario I we have now a metric in a signle column and a column with actual value

2. **Understanding the preservation**: There are fighter attributes such as teh name and the age the height they will continue remain in the indeifiter column and basically what it does is that it iwll allow us to give room to group and will I would be able to filter these going by the characteristics.

3. **The observation provided**: Ear row wil now have a speific measurement for only fighter insteading have all metric that is provided for that one fighter

the tidy format will basically contain the 100 rows in stead just have the original 10 rows that is being provided and wit htis method we will have only a fewer column to go by and this structure will basically allow us for more a flexible analysis of course and working with visualization

```python
# We are going to basically categorize metrics into groups 

metric_categories = {
    'record': ['wins', 'losses'],
    'striking': ['SLpM', 'sig_str_acc', 'SApM', 'str_def'],
    'grappling': ['td_avg', 'td_acc', 'td_def'],
    'submission': ['sub_avg']
}

# now after what is going to be ocured is that we are going to try to add a column of category to the tidy DF (DataFrame)

def categorize_metric(metric):
    for category, metrics in metric_categories.items():
        if metric in metrics:
            return category
    return 'other'

df_tidy['category'] = df_tidy['metric'].apply(categorize_metric)
```

This will basically enhance our dataset

1. **We are now going to be group the metrics**: Basically meaning adding a category that would be part of the each metric that is provided so we have the record, striking, grappling, and the submission

2. **Enabling the analysis and understanding**: What this will do for is that it will make understand the pattern and basically giving us a review of the fighting skills that is provided

3. **A visualization**: This will help overall it's going to be easier to create a visualization and we are going to be comparting the aspects of the fighter performance and the attribute

With the tidy structure we see that we are ready to understand the patterns with such more comprehnsive detail that I will be presenting

```python
# What we will be doing here is that we will be creating a corerelation when it comes down to the attributes of physical and undersanding the key metrics when it comes down to the perofrmance

key_metrics = ['wins', 'losses', 'SLpM', 'sig_str_acc', 'td_avg', 'sub_avg']
physical_attrs = ['height', 'weight', 'reach', 'age']

# In this step what I will be doing is basically we are going to need to use the wide dataset work for the correlation analysis taht we are going to be presented 

correlation_df = df[physical_attrs + key_metrics].corr()

# Now we are going to display the matrix of the correlation 

print("\nCorrelation between physical attributes and performance metrics:")
correlation_df

# now we are going to vusalize the matrix of the correlation

plt.figure(figsize=(10, 8))
plt.imshow(correlation_df, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation')
plt.title('Correlation Between Physical Attributes and Performance')

# we are going to add the correlation values as the text over here so:

for i in range(len(correlation_df.columns)):
    for j in range(len(correlation_df.index)):
        plt.text(i, j, f'{correlation_df.iloc[j, i]:.2f}', 
                 ha='center', va='center', color='black')

# we will be setting the ticks and the labels that is presenting in the work 

plt.xticks(range(len(correlation_df.columns)), correlation_df.columns, rotation=90)
plt.yticks(range(len(correlation_df.index)), correlation_df.index)
plt.tight_layout()
plt.show()
```

### Analysis 3: The win rate and the fighting style

Now what I will be doing is revieiwng the win rate that is related to style and understanding the performance that is being presented. 

```python
# we will be calcualting the percentage for each figther in terms of wins and loses to give us a better idea
df['win_percentage'] = df['wins'] / (df['wins'] + df['losses']) * 100

# we are going to be grouping to see the stance and the avg of the win percentage
stance_win_analysis = df.groupby('stance')['win_percentage'].mean().reset_index()
print("\nAverage win percentage by fighting stance:")
print(stance_win_analysis)

# we are going to visualize the stance win 
plt.figure(figsize=(10, 6))
plt.bar(stance_win_analysis['stance'], stance_win_analysis['win_percentage'], color='skyblue')
plt.title('Average Win Percentage by Fighting Stance', fontsize=14)
plt.xlabel('Stance', fontsize=12)
plt.ylabel('Win Percentage (%)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Analysis 4: Now we will take a look and view the performance of the weight class

Fianlly, at the final step we want to see the metrics of the performance and the difference across the weight that is presetned: 

```python
# what will be done is now we are going to group the data with the weigh class and the metric togehter
weight_performance = df_tidy.groupby(['weight_class', 'metric'], observed=True)['value'].mean().reset_index()

# we will have to filter and ability to focus on the main highlight of the metrics performance
key_performance = ['SLpM', 'SApM', 'td_avg', 'sub_avg']
weight_key_metrics = weight_performance[weight_performance['metric'].isin(key_performance)]

# for the next step we will officially be creating a visualization for each of the metrics and of course the weigh classes as well to give us a better idea 


plt.figure(figsize=(14, 10))

for i, metric in enumerate(key_performance):
    plt.subplot(2, 2, i+1)
    metric_data = weight_key_metrics[weight_key_metrics['metric'] == metric]
    
    # so now when it comes down ot the metrics what we will be doing here is sorting the weight class from the lightest weight to the heaviest weight
    metric_data = metric_data.sort_values('weight_class')
    
    plt.bar(metric_data['weight_class'], metric_data['value'], color='cornflowerblue')
    plt.title(f'Average {metric} by Weight Class', fontsize=12)
    plt.xlabel('Weight Class', fontsize=10)
    plt.ylabel('Value', fontsize=10)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('metrics_by_weight.png')  # we will once agian save the figure 
plt.show()
```
